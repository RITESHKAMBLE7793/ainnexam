<h3>Map Coloring Problem</h3>

```python
# Check if color can be assigned safely
def is_safe(node, color, graph, colors):

    # Check neighboring nodes
    for neighbor in graph[node]:

        # Return False if same color found
        if colors[neighbor] == color:
            return False

    # Return True if color is safe
    return True


# Graph Coloring Function
def graph_coloring(graph, m, colors, node=0):

    # Stop if all nodes are colored
    if node == len(graph):
        return True

    # Try colors from 1 to m
    for color in range(1, m + 1):

        # Check safe color
        if is_safe(node, color, graph, colors):

            # Assign color to node
            colors[node] = color

            # Recursive call for next node
            if graph_coloring(graph, m, colors, node + 1):
                return True

            # Backtracking step
            colors[node] = 0

    # Return False if no color possible
    return False


# Graph Representation
graph = {
    0: [1, 2, 3],
    1: [0, 2],
    2: [0, 1, 3],
    3: [0, 2]
}

# Number of colors
m = 3

# Store colors of nodes
colors = [0] * len(graph)

# Run Graph Coloring Algorithm
if graph_coloring(graph, m, colors):

    # Print solution
    print("Graph Coloring Solution:\n")

    for node, color in enumerate(colors):
        print(f"Node {node} -> Color {color}")

else:
    # Print if no solution exists
    print("No Solution Found")
```

<h3>Sudoku Solver</h3>

```python
# Print Sudoku Board
def print_board(board):

    # Loop through all rows
    for i in range(9):

        # Print horizontal line after every 3 rows
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - -")

        # Loop through all columns
        for j in range(9):

            # Print vertical line after every 3 columns
            if j % 3 == 0 and j != 0:
                print("| ", end="")

            # Print last element with new line
            if j == 8:
                print(board[i][j])

            # Print remaining elements
            else:
                print(board[i][j], end=" ")


# Find Empty Cell
def find_empty(board):

    # Search empty position in board
    for i in range(9):
        for j in range(9):

            # Empty cell contains 0
            if board[i][j] == 0:
                return (i, j)

    # Return None if board is full
    return None


# Check Valid Number
def valid(board, num, pos):

    row, col = pos

    # Check row
    for j in range(9):
        if board[row][j] == num:
            return False

    # Check column
    for i in range(9):
        if board[i][col] == num:
            return False

    # Find 3x3 box position
    box_x = col // 3
    box_y = row // 3

    # Check 3x3 box
    for i in range(box_y * 3, box_y * 3 + 3):
        for j in range(box_x * 3, box_x * 3 + 3):

            if board[i][j] == num:
                return False

    # Return True if number is valid
    return True


# Solve Sudoku Using Backtracking
def solve(board):

    # Find empty position
    empty = find_empty(board)

    # Stop if no empty cell found
    if not empty:
        return True

    row, col = empty

    # Try numbers from 1 to 9
    for num in range(1, 10):

        # Check valid number
        if valid(board, num, (row, col)):

            # Place number
            board[row][col] = num

            # Recursive call
            if solve(board):
                return True

            # Backtrack if solution fails
            board[row][col] = 0

    # Return False if no solution found
    return False


# Sudoku Board
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

# Print Initial Board
print("Initial Sudoku:\n")
print_board(board)

# Solve Sudoku
solve(board)

# Print Solved Board
print("\nSolved Sudoku:\n")
print_board(board)
```
