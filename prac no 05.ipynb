```python
import numpy as np
import matplotlib.pyplot as plt
```

# Cell 2 : Dataset

```python
# House Size and Rooms
X = np.array([
    [1000, 2],
    [1500, 3],
    [2000, 4],
    [2500, 4]
])

# House Prices
y = np.array([200, 300, 400, 500])
```

# Cell 3 : Feature Scaling

```python
# Normalize Data
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Add Bias Column
X = np.c_[np.ones(X.shape[0]), X]
```

# Cell 4 : Cost Function

```python
# Calculate Cost
def cost(X, y, theta):

    m = len(y)

    prediction = X.dot(theta)

    error = prediction - y

    return (1/(2*m)) * np.sum(error**2)
```

# Cell 5 : Batch Gradient Descent

```python
# BGD Algorithm
def BGD(X, y, lr=0.01, epochs=100):

    m, n = X.shape

    theta = np.zeros(n)

    history = []

    for i in range(epochs):

        prediction = X.dot(theta)

        error = prediction - y

        theta = theta - (lr/m) * X.T.dot(error)

        history.append(cost(X, y, theta))

    return theta, history
```

# Cell 6 : Stochastic Gradient Descent

```python
# SGD Algorithm
def SGD(X, y, lr=0.01, epochs=100):

    m, n = X.shape

    theta = np.zeros(n)

    history = []

    for j in range(epochs):

        for i in range(m):

            xi = X[i:i+1]

            yi = y[i:i+1]

            prediction = xi.dot(theta)

            error = prediction - yi

            theta = theta - lr * xi.T.dot(error)

        history.append(cost(X, y, theta))

    return theta, history
```

# Cell 7 : Mini Batch Gradient Descent

```python
# MBGD Algorithm
def MBGD(X, y, batch=2, lr=0.01, epochs=100):

    m, n = X.shape

    theta = np.zeros(n)

    history = []

    for j in range(epochs):

        for i in range(0, m, batch):

            xi = X[i:i+batch]

            yi = y[i:i+batch]

            prediction = xi.dot(theta)

            error = prediction - yi

            theta = theta - (lr/len(xi)) * xi.T.dot(error)

        history.append(cost(X, y, theta))

    return theta, history
```

```python
theta_bgd, cost_bgd = BGD(X, y)

theta_sgd, cost_sgd = SGD(X, y)

theta_mbgd, cost_mbgd = MBGD(X, y)
```
