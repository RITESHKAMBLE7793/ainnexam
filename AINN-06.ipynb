<h4>Handwritten Digit Recognition (MNIST Dataset)-Implement a 3-layer neural network
(Input → Hidden → Output) using Backpropagation.
    
1. Use Sigmoid, ReLU, or Tanh activation functions.
2. Train the network on the MNIST dataset to classify handwritten digits (0-9).
3. Optimize using Gradient Descent with Momentum or Adam Optimizer.</h4>

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
x_test.shape
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train.shape
x_test.shape
```

```python
index=0
plt.imshow(x_train[index])
print(y_train[index])
```

<h3>Relu Activation Function</h3>

```python
# Create Neural Network Model
model1 = Sequential([

    # Convert 2D image into 1D vector
    Flatten(input_shape=(28, 28)),

    # Hidden Layer with ReLU Activation
    Dense(128, activation='relu'),

    # Output Layer with Softmax Activation
    Dense(10, activation='softmax')

])
```

```python
# Compile the Model
model1.compile(

    # Optimization Algorithm
    optimizer=Adam(),

    # Loss Function
    loss='sparse_categorical_crossentropy',

    # Evaluation Metric
    metrics=['accuracy']

)
```

```python
history = model1.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model1.evaluate(x_test, y_test)
print(f"\nTest Accuracy (Relu): {test_acc:.4f}")
```

```python
plt.figure(figsize=(12,4))

# Accuracy Plot
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Test')
plt.title('Accuracy')
plt.legend()

# Loss Plot
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Test')
plt.title('Loss')
plt.legend()

plt.show()
```

<h3>Sigmoid Activation Function</h3>

```python
# Create Neural Network Model
model2 = Sequential([

    # Convert 2D image into 1D vector
    Flatten(input_shape=(28,28)),

    # Hidden Layer with Sigmoid Activation
    Dense(128, activation='sigmoid'),

    # Output Layer with Softmax Activation
    Dense(10, activation='softmax')

])
```

```python
# Compile the Model
model2.compile(

    # Optimization Algorithm
    optimizer=Adam(),

    # Loss Function
    loss='sparse_categorical_crossentropy',

    # Evaluation Metric
    metrics=['accuracy']

)
```

```python
history2 = model2.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model2.evaluate(x_test, y_test)
print(f"Test Accuracy (Sigmoid): {test_acc:.4f}")
```

```python
plt.figure(figsize=(12,4))

# Accuracy Plot
plt.subplot(1,2,1)
plt.plot(history2.history['accuracy'], label='Train')
plt.plot(history2.history['val_accuracy'], label='Test')
plt.title('Accuracy - Sigmoid')
plt.legend()

# Loss Plot
plt.subplot(1,2,2)
plt.plot(history2.history['loss'], label='Train')
plt.plot(history2.history['val_loss'], label='Test')
plt.title('Loss - Sigmoid')
plt.legend()

plt.show()
```

<h3>Tanh Activation Function</h3>

```python
# Create Neural Network Model
model3 = Sequential([

    # Convert 2D image into 1D vector
    Flatten(input_shape=(28,28)),

    # Hidden Layer with Tanh Activation
    Dense(128, activation='tanh'),

    # Output Layer with Softmax Activation
    Dense(10, activation='softmax')

])
```

```python
model3.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history3 = model3.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Evaluate Model on Test Data
test_loss, test_acc = model3.evaluate(x_test, y_test)

# Print Test Accuracy
print(f"Test Accuracy (Tanh): {test_acc:.4f}")
```

```python
plt.figure(figsize=(12,4))

# Accuracy Plot
plt.subplot(1,2,1)
plt.plot(history3.history['accuracy'], label='Train')
plt.plot(history3.history['val_accuracy'], label='Test')
plt.title('Accuracy - Tanh')
plt.legend()

# Loss Plot
plt.subplot(1,2,2)
plt.plot(history3.history['loss'], label='Train')
plt.plot(history3.history['val_loss'], label='Test')
plt.title('Loss - Tanh')
plt.legend()

plt.show()
```
