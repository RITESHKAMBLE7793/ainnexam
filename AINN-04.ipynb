<h3>First-Order Logic (FOL)</h3>

```python
# Define Rules
rules = {

    # If Fever and Cough then Flu
    "FeverAndCough_implies_Flu":
    ([("Fever", "x"), ("Cough", "x")], ("Flu", "x")),

    # If Flu then NeedsRest
    "Flu_implies_NeedsRest":
    ([("Flu", "x")], ("NeedsRest", "x")),

    # If SevereCough and ShortnessOfBreath then Pneumonia
    "SevereCoughAndBreath_implies_Pneumonia":
    ([("SevereCough", "x"), ("ShortnessOfBreath", "x")],
     ("Pneumonia", "x"))
}


# Initial Facts
facts = {
    ("Fever", "John"),
    ("Cough", "John")
}


# Store Proof Information
proofs = {f: "ASSERTED" for f in facts}


# Replace variable x with actual constant
def substitute(term, const):

    return tuple(const if t == "x" else t for t in term)


# Forward Chaining Function
def forward_chain():

    iteration = 1
    changed = True

    # Repeat until no new facts added
    while changed:

        print(f"\nIteration {iteration}")

        changed = False

        # Check every rule
        for rule, (premises, conclusion) in rules.items():

            # Get all constants from facts
            for const in {c for _, c in facts}:

                # Replace x with constant
                prem = [substitute(p, const) for p in premises]
                concl = substitute(conclusion, const)

                # Check if all premises exist
                if all(p in facts for p in prem) and concl not in facts:

                    # Add new fact
                    facts.add(concl)

                    # Store proof
                    proofs[concl] = (rule, prem)

                    print(f"Inferred: {concl}")

                    changed = True

        iteration += 1


# Explain Query Function
def explain(query):

    print(f"\nQuery: {query}")

    # Check if query exists
    if query not in facts:

        print("Fact NOT Known")
        return

    # Check asserted fact
    if proofs[query] == "ASSERTED":

        print("Asserted Fact")

    else:

        # Get proof details
        rule, premises = proofs[query]

        print(f"Inferred using rule: {rule}")

        print("Support Facts:")

        # Explain supporting facts
        for p in premises:
            explain(p)


# Run Forward Chaining
forward_chain()


# Query Explanations
explain(("Flu", "John"))

explain(("NeedsRest", "John"))

explain(("Pneumonia", "John"))
```
