<h4>Develop a knowledge base for a medical diagnosis system using First Order Logic predicates
such as:
    
1. HasSymptom(John, Flu)
2. HasDisease(Kento, diarrhea)</h4>

<h3>First-Order Logic predicates</h3>

```python
facts = {("Fever", "John"), ("Cough", "John")}

rules = [
    ([("Fever", "John"), ("Cough", "John")], ("Flu", "John")),
    ([("Flu", "John")], ("NeedsRest", "John")),
    ([("SevereCough", "John"), ("ShortnessOfBreath", "John")], ("Pneumonia", "John"))
]

changed = True
while changed:
    changed = False
    for premises, conclusion in rules:
        if all(p in facts for p in premises) and conclusion not in facts:
            facts.add(conclusion)
            changed = True

queries = ["Flu", "NeedsRest", "Pneumonia"]
for disease in queries:
    if (disease, "John") in facts:
        print(f"{disease}(John) is True")
    else:
        print(f"{disease}(John) is False")
```
